* #### Function with unlimited auguments

```python
def sum(*a):
    sum = 0  # 10, 89, 10, 89, 78, 14, 15
    for i in a:
        sum = i + sum
    
    return sum

sum(10, 89, 10, 89, 78, 14, 15)
```

* #### Function with unlimited keyword auguments

```python
def sum(**a):
    print(a)

sum(a = 5, b = 5, c = 6, d = 10, e = 67)
```

* #### Lamda functions

```python
# def square(a):
#     return a**2

square = lambda a, b: a + b

square(4, 5)
```

* #### Recursive functions


```python
def name(a):
    print('----------------')
    print(a)
    name(a)

name('Kamran')
```

### Loops

```python
for i in range(5):
    for j in ['apple', 'banana', 'cherry']:
        print(f'{i}- {j}')
    print()
```

```python
for i in range(10):
    print(i)
    if i == 5:
        break
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```
